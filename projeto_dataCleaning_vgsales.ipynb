```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/vgsales.csv')
dados.head(10)
```

```python
# pegando serie JP_Sales separando para análises
dados_jp = dados['JP_Sales']
dados_s_jp = pd.DataFrame(dados_jp)
```

```python
dados_s_jp.mask(dados_s_jp == 0.00, 0.01)
```

```python
# transformando todos valores "0.00" das colunas "Sales" em valores "0.01"
new_dados = dados.mask(dados == 0.00, 0.01)
new_dados.tail(20)
```

```python
new_dados.info()
```

```python
# verificando valores null ou NaN
check_na = new_dados.isna()
check_na.value_counts()
```

```python
# vericando valores null ou NaN e fazendo a contagem
new_dados.isnull().sum()
```

```python
# excluindo valores nulos das series "Year" e "Publisher" 
new_dados.dropna(axis=0, inplace=True)
```

```python
new_dados.isnull().sum()
```

```python
# transformando a serie "Rank" em index
new_dados.set_index('Rank', inplace=True)
```

```python
# renomeando colunas
new_dados.rename(columns={"Name":"name",
                  "Platform":"plataform",
                  "Year":"year", "Genre":"genre", 
                  "Publisher":"publisher", 
                  "NA_Sales":"na_sales", 
                  "EU_Sales":"eu_sales", 
                  "JP_Sales":"jp_sales", 
                  "Other_Sales":"other_sales",
                  "Global_Sales":"global_sales"}, inplace=True)
```

```python
# convertendo tipo de dado
new_dados['year'] = new_dados['year'].astype(int)
```

```python
new_dados.head(10)
```

```python
# retirando caracter dos tipos de dados de uma coluna
new_dados["name"] = new_dados["name"].apply(lambda x: x.replace(".", ""))
```

```python
new_dados['name'].head(20)
```

```python
# expotando o novo dataset
new_dados.to_csv('dados/new_vgsales.csv')
```
